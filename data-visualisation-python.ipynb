```python
data = pd.read_csv('countries.csv')
```

```python
# data.country.unique()
us = data[data.country=='United States']
china = data[data.country == 'China']
```

```python
us.head()
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population/10**6)
plt.xlabel('year')
plt.ylabel('population in millions')
plt.title('Population of Us and China')
plt.legend(['us', 'china'])
```

```python
# print(type(data.population), data.population[0])
# print(type(us.population), us.population[0])
'''
for position based indexing use iloc[x]
for label based indexing use [label]
we can use position based indexing for [x] but the dataframe has to be sequential or integer based otherwise face the error
'''
```

```python
plt.plot(us.year, (us.population / us.population.iloc[0]) * 100)
plt.plot(china.year, (china.population / china.population.iloc[0]) * 100)
plt.xlabel('year')
plt.ylabel('population percentage change')
plt.title('Population of Us and China')
plt.legend(['us', 'china'])
```
